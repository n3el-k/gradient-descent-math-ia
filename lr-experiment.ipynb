```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

class SimpleModel:
    def __init__(self):
        self.W1 = np.array([[0.5, 0.2], [0.3, 0.8]]) # 2x2
        self.b1 = np.zeros((1, 2)) # 1x2
        self.W2 = np.array([[0.6, 0.4]]) # 1x2
        self.b2 = np.zeros((1, 1)) # 1x1
        self.lr = 50
        self.y_hat = 0
        self.x = np.zeros((1,2))

    def forward(self, x):
        self.x = x # (1x2)
        self.z1 = x @ self.W1 + self.b1 # (1x2) x (2x2) = (1x2)
        self.h1 = sigmoid(self.z1)  # (1x2)
        self.z2 = self.h1 @ self.W2.T + self.b2 # (1x2) x (1x2)T = (1x1)
        self.y_hat = sigmoid(self.z2) # 1x1

        return self.y_hat
    
    def updateState(self, y):
        # delta = dL/dz2
        delta = 2 * self.lr * (y - self.y_hat) * self.y_hat * (1 - self.y_hat)
        self.W1 = self.W1 + delta * self.W2.T * self.h1 * (1 - self.h1) * self.x.T
        self.b1 = self.b1 + delta * self.W2 * self.h1 * (1 - self.h1)
        self.W2 = self.W2 + delta * self.h1
        self.b2 = self.b2 + delta
    
    def changeLr(self, lr):
        """For the lr experiment"""
        self.lr = lr
```

```python
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])
model = SimpleModel()
num_epochs = 100000

def train(num_epochs, lr):
    model.changeLr(lr)
    for epoch in range(num_epochs):
        total_loss = np.float64(0)
        for i in range(len(X)):
            x_i = X[i:i+1]
            y_i = y[i:i+1]
            pred = model.forward(x_i)
            model.updateState(y_i)
            total_loss += (y_i - pred)**2

        if epoch % 10000 == 0:
            print(f"Epoch {epoch}, Loss: {total_loss.item():.4f}")
```

```python
train(num_epochs, 0.01)
```

```python
train(num_epochs, 0.1)
```

```python
train(num_epochs, 10)
```

```python
train(num_epochs, 100)
```
